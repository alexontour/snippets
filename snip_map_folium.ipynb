```python
# https://python-visualization.github.io/folium/quickstart.html
m = folium.Map(location=[47.8,13.0], zoom_start=10)

tooltip = "ToolTip"

folium.Marker([47.8, 13.1], popup="<i>Test PopUp</i>", tooltip=tooltip).add_to(m)
folium.Marker(location=[47.8, 13.2], popup="Test Popup mit Icon", icon=folium.Icon(icon="cloud"), tooltip=tooltip).add_to(m)
folium.Marker(location=[47.8, 13.3], popup="Test Popup mit Farbe", icon=folium.Icon(color="green", icon="cloud"), tooltip=tooltip).add_to(m)

m
```

Zusammenfassen von Markern in Cluster

```python
# https://github.com/python-visualization/folium/tree/main/examples 

import folium
from folium.plugins import MarkerCluster


m = folium.Map(location=[47.8, 13.0], zoom_start=5)

marker_cluster = MarkerCluster().add_to(m)


folium.Marker(
    location=[47.8, 13.1],
    popup="Add popup text here.",
    icon=folium.Icon(color="green", icon="ok-sign"),
).add_to(marker_cluster)

folium.Marker(
    location=[47.8, 13.2],
    popup="Add popup text here.",
    icon=folium.Icon(color="red", icon="remove-sign"),
).add_to(marker_cluster)

folium.Marker(
    location=[47.8, 13.3],
    popup="Add popup text here.",
    icon=None,
).add_to(marker_cluster)

m
```

```python
from folium import FeatureGroup, LayerControl, Map, Marker
from folium.plugins import MarkerCluster
from folium import plugins


m = Map(location=[47.8, 13.0], zoom_start=9, tiles="OpenStreetMap")

#tiles="Stamen Terrain",
#tiles="Stamen Toner",
#tiles="Stamen Watercolor",
#tiles="Mapbox Bright",
#tiles="Mapbox Control Room",
#tiles="OpenStreetMap" (default)
#tiles="CartoDB Positron",
#tiles="CartoDB Dark_Matter",


feature_group_1 = FeatureGroup(name="Gruppe 1")
marker_cluster_1 = MarkerCluster().add_to(feature_group_1)
Marker(location=[47.8, 13.1], popup="Marker 1", icon=folium.Icon(color="red", icon="hospital")).add_to(marker_cluster_1)
Marker(location=[47.8, 13.2], popup="Marker 2", icon=folium.Icon(color="red", icon="bolt", prefix="fa")).add_to(marker_cluster_1)

feature_group_2 = FeatureGroup(name="Gruppe 2")
marker_cluster_2 = MarkerCluster().add_to(feature_group_2)
Marker(location=[47.8, 13.3], popup="Marker 3", icon=folium.Icon(color="blue", icon="remove-sign", prefix="glyphicon")).add_to(marker_cluster_2)
Marker(location=[47.8, 13.4], popup="Marker 4", icon=folium.Icon(color="blue", icon="glyphicon-stats", prefix="glyphicon")).add_to(marker_cluster_2)

feature_group_1.add_to(m)
feature_group_2.add_to(m)
LayerControl().add_to(m)

minimap = plugins.MiniMap(toggle_display=True)
m.add_child(minimap)
plugins.ScrollZoomToggler().add_to(m)
plugins.Fullscreen(position="topright").add_to(m)

m
```

```python
m.save("folium_map.html")
```
