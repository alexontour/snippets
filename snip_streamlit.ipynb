```python
import pandas as pd
import streamlit as st
```

```python
df = pd.DataFrame({'val': [1, 2, 3], 'key': ['Item1List1', 'Item1List1', 'Item2List1']})
```

```python
list1 = ['Item1List1', 'Item2List1']
```

```python
list1_choice = st.sidebar.multiselect(
    'List-1:', list1, default=list1)
```

```python
slider_choice = st.sidebar.slider(
    'Slider:', min_value=4, max_value=15.0, step=.5, value=7.0)
```

```python
# Filter df
df = df[df['key'].isin(list1_choice)]
#df = df[df['team'].isin(teams_choice)]
#df = df[df['cost'] < price_choice]
```

```python
st.title(f"Demo")
```

```python
st.dataframe(df.sort_values('val',
             ascending=False).reset_index(drop=True))
```
