```python
# importieren der notwendigen Bibliotheken

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

# Daten laden

```python
# CSV in Dataframe einlesen

url = 'https://www.salzburg.gv.at/ogd/bad388c1-e13f-484d-ba51-331a79537f5f/meteorologie-aktuell.csv'
df = pd.read_csv(url, encoding='ISO-8859-1', sep=';')
```

```python
# Dataframe anzeigen (die ersten 5 Zeilen)

#df.head()
df
```

```python
# Zusammenfasende Statistik

df.describe()
```

```python
# Datentypen des Dataframe anzeigen

df.dtypes
```

# Daten bereinigen

```python
# Datentypen richtig stellen

df = df.convert_dtypes()
df.replace('?', np.nan, inplace=True)
df['Messwert'] = df['HMW'].str.replace(',', '.').astype(float)
df['DateTime'] = pd.to_datetime(df['Zeitpunkt'], format='%d.%m.%Y %H:%M', dayfirst=True)
```

```python
# Datentypen des Dataframe anzeigen
df.dtypes
```

```python
df.info()
```

```python
# Dataframe anzeigen
df
```

# Daten filtern

```python
# Einzelwerte für den Filter darstellen

unique_values = df['Messort'].unique()
num_unique_values = df['Messort'].nunique()

# Ergebnis ausgeben
print("Eindeutige Werte:")
print(unique_values)

print("\nAnzahl der eindeutigen Werte:", num_unique_values)
```

```python
# nach zeilen filtern

df_filtered = df.query('Messort == "Salzburg Flughafen" & Parameter == "Lufttemperatur [GradC]"')
```

```python
df_filtered.head()
```

```python
# Zusammenfasende Statistik
df_filtered.describe()
```

# Daten darstellen

```python
# Einen einfache Linienplot erstellen
plt.plot(df_filtered['DateTime'], df_filtered['Messwert'])

# Datetime-Format für die x-Achse festlegen (optional)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
# Beschriftung schräg stellen (optional)
plt.xticks(rotation=45)

# Achsenbeschriftung
plt.xlabel('DateTime')
plt.ylabel('Parameter')
plt.title('Zeitserie')



# Plot darstellen
plt.show()
```
