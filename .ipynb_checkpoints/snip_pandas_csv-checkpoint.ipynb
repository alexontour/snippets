```python
# importieren der notwendigen Bibliotheken

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# CSV in Dataframe einlesen

url = 'https://www.salzburg.gv.at/ogd/bad388c1-e13f-484d-ba51-331a79537f5f/meteorologie-aktuell.csv'
df = pd.read_csv(url, encoding='ISO-8859-1', sep=';')
```

```python
# Dataframe anzeigen und die ersten 5 Zeilen anzeigen

df.head()
```

```python
# Zusammenfasende Statistik - Achtung - macht nicht immer Sinn !
df.describe()
```

```python
# Datentypen des Dataframe anzeigen
df.dtypes
```

```python
# Datentypen richtig stellen
df = df.convert_dtypes()
df.replace('?', np.nan, inplace=True)
df['HMW'] = df['HMW'].str.replace(',', '.').astype(float)
df['DateTime'] = pd.to_datetime(df['Zeitpunkt'], format='%d.%m.%Y %H:%M', dayfirst=True)
```

```python
# Datentypen des Dataframe anzeigen
df.dtypes
```

```python
df.info()
```

```python
# Dataframe anzeigen
df
```

```python
# einzelwerte darstellen

unique_values = df['Messort'].unique()
num_unique_values = df['Messort'].nunique()

# Ergebnis ausgeben
print("Eindeutige Werte:")
print(unique_values)

print("\nAnzahl der eindeutigen Werte:", num_unique_values)
```

```python
# nach zeilen filtern

df_filtered = df.query('Messort == "Salzburg Flughafen" & Parameter == "Lufttemperatur [GradC]"')
```

```python
df_filtered
```

```python
# Zusammenfasende Statistik
df_filtered.describe()
```

```python
# Initialisierung des Plots
fig, ax = plt.subplots()

# Daten f√ºr den Plot ausw√§hlen
ax.plot(df_filtered['DateTime'], df_filtered['HMW'], marker='o', linestyle='-')

# Achsenbeschriftung
plt.xlabel('DateTime')
plt.ylabel('Parameter')
plt.title('Zeitserie')

# Beschriftung schr√§g stellen (optional)
plt.xticks(rotation=45)

# Plot darstellen
plt.show()
```
