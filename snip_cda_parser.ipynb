```python
# Read from Gitlab URL
# CDA - AMB 
url = 'https://gitlab.com/elga-gmbh/cda-beispielbefunde/-/raw/master/Ambulanzbefund/Allgemeiner_Ambulanzbefund_-_codiert.xml'
contents = urllib.request.urlopen(url).read()
#contents
```

```python
# CDA parsen mit BeautifulSoup
etree = ET.fromstring(contents)
for docTitle in etree.findall('{urn:hl7-org:v3}title'):
  print(docTitle.text)
for compTitle in etree.findall('.//{urn:hl7-org:v3}title'):
  print(compTitle.text)
```

```python
# CDA parsen mit BeautifulSoup
soup = BeautifulSoup(contents,'xml')
titles = soup.find_all('title')
for title in titles:
    print(title.get_text())
```
